```python
import pandas as pd
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')
```

```python
#3
import numpy as np

def q1(x): return x.quantile(0.25)
def q3(x): return x.quantile(0.75)

pivot_cau3 = pd.pivot_table(
    df,
    values='DH1',
    index='KT',
    aggfunc=['count', 'sum', 'mean', 'median', 'min', 'max', 'std', q1, q3]
)

pivot_cau3.columns = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 'Q1', 'Q3']
print(pivot_cau3)
```

```python
#4
pivot_cau4 = pd.pivot_table(
    df,
    values='DH1',
    index=['KT', 'KV'],
    aggfunc=['count', 'sum', 'mean', 'median', 'min', 'max', 'std', q1, q3]
)

pivot_cau4.columns = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 'Q1', 'Q3']
print(pivot_cau4)
```

```python
#5
import pandas as pd
import numpy as np

def q1(x): return x.quantile(0.25)
def q3(x): return x.quantile(0.75)

pivot_cau5 = pd.pivot_table(
    df,
    values='DH1',
    index=['KT', 'KV', 'DT'],
    aggfunc=['count', 'sum', 'mean', 'median', 'min', 'max', 'std', q1, q3]
)

pivot_cau5.columns = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 'Q1', 'Q3']

print(pivot_cau5)
```

Phần 2: Trình bày dữ liệu

```python
#1
tan_so = df['GT'].value_counts()

tan_suat = df['GT'].value_counts(normalize=True) * 100

bang_gt = pd.DataFrame({'Tần số': tan_so, 'Tần suất (%)': tan_suat})
print(bang_gt)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 1. Biểu đồ tần số (Cột)
sns.countplot(data=df, x='GT', ax=axes[0], palette='Set2')
axes[0].set_title('Biểu đồ tần số biến GT')
axes[0].set_ylabel('Số lượng')
axes[0].set_xlabel('Giới tính')

# 2. Biểu đồ tần suất (Tròn)
axes[1].pie(tan_suat, labels=tan_suat.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
axes[1].set_title('Biểu đồ tần suất biến GT')

plt.tight_layout()
plt.show()
```

```python
#2
# 1. Bảng thống kê mô tả cho cả 3 biến
print(df[['US_TBM1', 'US_TBM2', 'US_TBM3']].describe())

# 2. Vẽ biểu đồ phân phối để so sánh
plt.figure(figsize=(10, 6))
for col in ['US_TBM1', 'US_TBM2', 'US_TBM3']:
    sns.kdeplot(df[col], shade=True, label=col)

plt.title('Biểu đồ phân phối của US_TBM1, US_TBM2 và US_TBM3')
plt.xlabel('Điểm số')
plt.ylabel('Mật độ')
plt.legend()
plt.show()
```

```python
#3

df_nam = df[df['GT'] == 'M']

# Lập bảng tần số và tần suất của biến DT
tan_so_dt_nam = df_nam['DT'].value_counts()
tan_suat_dt_nam = df_nam['DT'].value_counts(normalize=True) * 100

bang_dt_nam = pd.DataFrame({'Tần số': tan_so_dt_nam, 'Tần suất (%)': tan_suat_dt_nam})
print("Thống kê biến DT cho học sinh Nam:")
print(bang_dt_nam)

# Vẽ biểu đồ cột trực quan hóa
plt.figure(figsize=(8, 4))
sns.barplot(x=tan_so_dt_nam.index, y=tan_so_dt_nam.values, palette='Blues_r')
plt.title('Phân phối biến DT của các học sinh Nam')
plt.xlabel('Dân tộc')
plt.ylabel('Số lượng')
plt.show()
```

```python
#4
# Lọc theo các điều kiện kết hợp
cond_cau4 = (df['GT'] == 'M') & \
            (df['DT'] == 'Kinh') & \
            (df['DH1'] >= 5.0) & \
            (df['DH2'] >= 4.0) & \
            (df['DH3'] >= 4.0)

df_cau4 = df[cond_cau4]

# Trình bày biến KV dưới dạng bảng tần số, tần suất
tan_so_kv = df_cau4['KV'].value_counts()
tan_suat_kv = df_cau4['KV'].value_counts(normalize=True) * 100

bang_kv = pd.DataFrame({'Tần số': tan_so_kv, 'Tần suất (%)': tan_suat_kv})
print("Thống kê biến KV thỏa mãn các điều kiện câu 4:")
print(bang_kv)
```

```python
#5
cond_cau5 = (df['KV'] == '2NT') & (df['DH1'] >= 5.0) & (df['DH2'] >= 5.0) & (df['DH3'] >= 5.0)
df_cau5 = df[cond_cau5]

# Trình bày các thông số mô tả cơ bản của 3 biến điểm
print("Thống kê mô tả điểm DH1, DH2, DH3 thỏa mãn điều kiện câu 5:")
print(df_cau5[['DH1', 'DH2', 'DH3']].describe())

plt.figure(figsize=(8, 5))
sns.boxplot(data=df_cau5[['DH1', 'DH2', 'DH3']], palette='Pastel1')
plt.title('Biểu đồ Boxplot điểm DH1, DH2, DH3 (Thỏa mãn ĐK câu 5)')
plt.ylabel('Điểm số')
plt.show()
```

Phần 4: Trực quan hóa dữ liệu nâng cao

```python
#2&3
# Định nghĩa hàm phân loại
def phan_loai(x):
    if x < 5: return 'k'
    elif x < 7: return 'tb'
    elif x < 8: return 'k'
    else: return 'g'

df['phanlopt1'] = df['T1'].apply(phan_loai)

print(df['phanlopt1'].value_counts())
```

```python
#4&5
# Gom nhóm dữ liệu theo T1 và phanlopt1
data_grouped = df.groupby(['T1', 'phanlopt1']).size().unstack(fill_value=0)

# Vẽ Multiple Line
data_grouped.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Multiple Line: T1 theo phanlopt1')
plt.show()

# Vẽ Drop-line (Stem plot)
plt.figure(figsize=(10, 6))
for col in data_grouped.columns:
    plt.stem(data_grouped.index, data_grouped[col], label=col, basefmt=" ")
plt.legend()
plt.title('Drop-line: T1 theo phanlopt1')
plt.show()
```
