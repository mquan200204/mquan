```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Nạp dữ liệu
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

# Kiểm tra dữ liệu
print("Thông tin dữ liệu:")
print(df.info())
print("\n5 dòng đầu:")
print(df.head())
```

```python
# CÂU 3: TẢI DỮ LIỆU VÀ IN DÒNG ĐẦU/CUỐI
print("--- 10 DÒNG ĐẦU TIÊN ---")
print(df.head(10))

print("\n--- 10 DÒNG CUỐI CÙNG ---")
print(df.tail(10))
```

```python
# CÂU 4:
print(f"\nSố dòng thiếu ở cột DT trước xử lý: {df['DT'].isnull().sum()}")
print(f"Các giá trị riêng biệt của DT: {df['DT'].unique()}")

df['DT'] = df['DT'].fillna(0)
```

```python
# CÂU 5 & 6:
mean_t1 = df['T1'].mean()
df['T1'] = df['T1'].fillna(mean_t1)

score_columns = [
    'T1', 'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1',
    'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2',
    'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6',
    'DH1', 'DH2', 'DH3'
]
for col in score_columns:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mean())
```

```python
# CÂU 7:
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 1
```

```python
# CÂU 8:
bins = [0, 5.0, 6.5, 8.0, 9.0, 10.1]
labels = ['Y', 'TB', 'K', 'G', 'XS']

df['XL1'] = pd.cut(df['TBM1'], bins=bins, labels=labels, right=False)
df['XL2'] = pd.cut(df['TBM2'], bins=bins, labels=labels, right=False)
df['XL3'] = pd.cut(df['TBM3'], bins=bins, labels=labels, right=False)
```

```python
# CÂU 9:
df['US_TBM1'] = (df['TBM1'] / 10.0) * 4
df['US_TBM2'] = (df['TBM2'] / 10.0) * 4
df['US_TBM3'] = (df['TBM3'] / 10.0) * 4
```

```python
# CÂU 10:
def tinh_kqxt(row):
    khoi = row['KT']
    dh1, dh2, dh3 = row['DH1'], row['DH2'], row['DH3']

    if khoi in ['A', 'A1']:
        diem_tb = (dh1 * 2 + dh2 + dh3) / 4
    elif khoi == 'B':
        diem_tb = (dh1 + dh2 * 2 + dh3) / 4
    else:
        diem_tb = (dh1 + dh2 + dh3) / 3

    return 1 if diem_tb >= 5.0 else 0
df['KQXT'] = df.apply(tinh_kqxt, axis=1)
```

```python
# CÂU 11:
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
print("\n[THÀNH CÔNG] Đã tiền xử lý và lưu thành file 'processed_dulieuxettuyendaihoc.csv'!")
```
